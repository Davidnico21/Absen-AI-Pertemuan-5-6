```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings("ignore")
```

```python
# membaca file CSV dari URL dan menyimpannya ke dalam Dataframe df
df = pd.read_csv("https://raw.githubusercontent.com/rahmantaufik27/dataset/main/study/daily_weather_prediction.csv")
df
```

```python
# Menampilkan informasi tentang data, mulai dari jumlah baris dan kolom, tipe data, nilai non-null dan penggunaan memori
df.info()
```

```python
# meilihat jumlah nilai null di setiap kolom.
df.isnull().sum()
```

```python
# melihat ringkasan statistik deskriptif untuk setiap kolom dalam bentuk yang lebih mudah dibaca
df.describe().transpose()
```

##Preprocessing

```python
# menghapus baris yang mengandung nilai null dari dataframe df dan menyimpan kembali ke variabel df
df = df.dropna()
df
```

```python
#Menambahkan kolom baru 'high_humidity_label' ke dataframe df
df['high_humidity_label'] = (df['relative_humidity_3pm'] > 24.99) * 1
df
```

```python
#Mendapatkan list nama-nama kolom dari DataFrame df
features = list(df.columns)

#Menghapus beberapa kolom dari list fitur
features.remove('number')
features.remove('relative_humidity_9am')
features.remove('relative_humidity_3pm')

#Menampilkan list fitur yang telah dimodifikasi
features
```

```python
#Memilih subset dari DataFrame df yang hanya terdiri dari fitur-fitur yang telah dipersiapkan sebelumnya
x = df[features].copy()

#Memilih kolom 'high_humidity_label' dari DataFrame df sebagai target variabel yang akan diprediksi
y = df["high_humidity_label"].copy()
```

```python
train_pct_index = int(0.8 * len(x))
x_train, x_test = x[:train_pct_index], x[train_pct_index:]
y_train, y_test = y[:train_pct_index], y[train_pct_index:]
```

```python
#Membuat objek DecisionTreeClassifier dengan mengatur jumlah maksimum leaf nodes menjadi 10
dt_cls = DecisionTreeClassifier(max_leaf_nodes=10, random_state=0)

#Melatih model DecisionTreeClassifier menggunakan data pelatihan yang telah dipersiapkan
dt_cls.fit(x_train, y_train)
```

##Prediction

```python
#Melakukan prediksi menggunakan model DecisionTreeClassifier pada data pengujian x_test
y_pred = dt_cls.predict(x_test)
y_pred
```

```python
#Menghitung akurasi antara label yang sebenarnya (y_test) dan label yang diprediksi (y_pred)
acc = accuracy_score(y_test, y_pred)

#Menampilkan akurasi dalam bentuk persentase dengan dua angka desimal
print("Akurasi: %.2f%%" % (acc * 100))
```

##Conclusion
Dengan menerapkan model Decision Tree Classifier yang dibatasi hingga maksimum 10 leaf nodes, akurasi prediksi pada data uji mencapai sekitar 78.87%. Hasil ini menunjukkan bahwa model tersebut mampu memprediksi label "High_Humidity" dengan akurasi sekitar 78.87% menggunakan fitur-fitur yang telah dipilih.
